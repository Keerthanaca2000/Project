# Outlier Detection and Treatment

```python
fig = plt.figure(figsize =(5, 5))
df.boxplot(column = 'Oldpeak',color='blue');
```

```python
Q1 = np.percentile(df['Oldpeak'],25,interpolation = 'midpoint')
print('Q1 = ',Q1)
Q3 = np.percentile(df['Oldpeak'],75,interpolation = 'midpoint')
print('Q3 = ',Q3)
print('Median = ',((Q1 + Q3)/2))
```

```python
IQR =Q3 -Q1
print('IQR = ',IQR)

#upper bound
upper = Q3 + (1.5*IQR)
print('UPPER BOUND = ',upper)

#lower bound
lower = Q3 - (1.5*IQR)
print('LOWER BOUND = ',lower)
```

```python
new_data=df[data["Oldpeak"]>lower].dropna()
new_data=df[new_data["Oldpeak"]<upper].dropna()
```

```python
fig = plt.figure(figsize =(5, 5))
new_data.boxplot(column = 'Oldpeak',color='blue');
```

```python
fig = plt.figure(figsize =(100, 5))
sns.boxplot(data=df)
```

```python
numerical_feature = df.describe(include=["int", "float"]).columns

# Print it in list type ..
print(list(numerical_feature))
```

```python
for idx, col  in enumerate(numerical_feature, 1):
    plt.figure(figsize=(5,5))
    
#     plt.subplot(len(numerical_feature) // 2, 3, idx)
    plt.boxplot(df[col])

    plt.title(col)
#     plt.hist(forest_df[col])


plt.tight_layout()
plt.show(plt)
```

```python
z_scores = pd.DataFrame()
for col in numerical_feature:
    z_score = (df[col] - df[col].mean()) / df[col].std()
    z_scores[col] = z_score

# Define a threshold for Z-score (e.g., 3)
z_score_threshold = 2

# Identify the indices of outliers based on the Z-score threshold
outliers_idx = np.where(np.abs(z_scores) > z_score_threshold)

# Remove the outliers from the dataset
data_cleaned = df.drop(outliers_idx[0])

# Note that you may need to reset the index of the cleaned dataset using:
# data_cleaned.reset_index(drop=True, inplace=True)

# Now you can create boxplots for the cleaned dataset without outliers
for col in numerical_feature:
    plt.figure(figsize=(5,5))
    plt.boxplot(data_cleaned[col])
    plt.title(col)
    plt.show()
```
